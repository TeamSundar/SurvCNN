```python
import pandas as pd
from matplotlib import pyplot as plt
from upsetplot import UpSet
from upsetplot import from_contents
```

```python
diff_enrich = pd.read_csv('data/enrichr_results/tsne_mrna_PH_enrichr_keggPathway_neg.txt', sep="\t")
```

```python
dict = {}
for i in range(20):
    dict[diff_enrich['Term'][i]]=diff_enrich['Genes'][i].split(';')
a = from_contents(dict)
```

```python
a['P-value']=diff_enrich['P-value'][0:20]
```

```python
upset = UpSet(a, subset_size='auto', intersection_plot_elements=5,orientation='horizontal',element_size=20)
#upset.add_catplot(value='P-value', kind='strip', color='blue')
#upset.add_catplot(value='AGE', kind='strip', color='black')
upset.plot()
plt.show()
```
